```python
## Exercise 7:
```

```python
x = np.linspace(-10, 10, 1000)
y1 = x**-1 * np.sin(x)
y2 = x**2+np.exp(x)
```

```python
func_y1, axes_func = plt.subplots(1,2, figsize=(10,5))
axes_func[0].set(title='x^-1 * sin(x)', xlabel = 'X', ylabel = 'Y')
axes_func[0].plot(x, y1, color='tab:blue')
axes_func[1].set(title='x^2 + exp(x)', xlabel = 'X', ylabel = 'Y')
axes_func[1].scatter(x, y2, color='red', edgecolors='black', s=50)
plt.tight_layout()
```

```python
## Exercise 8:
```

```python
x = np.linspace(-10, 10, 1000)
y1 = x * np.sin(x)
y2 = x**2+np.exp(x)
```

```python
func, axes_set = plt.subplots(2,figsize=(10,10))

axes_set[0].set(title = 'x * sin(x)')
axes_set[0].plot(x, y1, color='blue')

axes_set[1].set(title = 'x^2 + exp(x)')
axes_set[1].scatter(x, y2, color='green', marker='H', s=50, edgecolors='red')
plt.tight_layout()
```

```python
## Exercise 9:
```

```python
x = np.linspace(-10, 10, 1000)
y1 = x**-1 * np.tan(x)
y2 = x**2/np.exp(x)
```

```python
func, axes_set = plt.subplots(2, figsize=(10,10))

axes_set[0].set(xlabel = 'X', ylabel = 'Y')
axes_set[0].plot(x, y1, color='blue', label='y = x^-1 * tan(x)')
axes_set[0].legend(title='function 1', loc='lower left', shadow=True)

axes_set[1].set(title = 'x**2+np.exp(x)', xlabel = 'X', ylabel = 'Y')
axes_set[1].scatter(x, y2, color='green', marker='H', s=50, edgecolors='red', label = 'y = x^2/exp(x)')
axes_set[1].legend(title='function 2', loc='upper right', shadow=True)

plt.tight_layout()
```
