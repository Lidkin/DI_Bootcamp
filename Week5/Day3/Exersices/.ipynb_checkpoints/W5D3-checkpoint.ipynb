```python
x = np.linspace(0,10,20)
y1 = x**2
y2 = x**1.5
figure, axes = plt.subplots()
axes.set(title='exercise 4', xlabel='X', ylabel='Y')
axes.plot(x,y1, color='red', linestyle='dashed', marker='o', label = 'y = x^2')
axes.plot(x,y2,color='blue', linestyle=':', marker='s', label = 'y = x^1.5')
axes.legend()

figure.savefig('my_styled_plot.png')
```

```python
x = np.linspace(-10,10,10)
y1 = x ** 2
y2 = x ** 3

figure, axes_set = plt.subplots(2)

axes_set[0].plot(x, y1, color = 'red', label='main 1 line')
axes_set[0].annotate('breakout!', xy=(0,0), xytext=(1,10), arrowprops={'facecolor':'black', 'shrink':200})
axes_set[0].legend(title='Legend', loc='lower right', shadow=True)

axes_set[1].plot(x, y2, color = 'green', label='main 2 line')
axes_set[1].legend()
```

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

fig = plt.figure()
ax = plt.axes(projection='3d')

z_line = np.sin(x ** 2)**2
x_line = np.linspace(0, 2 * np.pi, 400)
y_line = np.sin(x ** 2)

ax.plot3D(x_line, y_line, z_line, 'gray')

z_points = y**2
x_points = x
y_points = y

ax.scatter3D(x_points, y_points, z_points, c=z_points, cmap='hsv')
```
