```python
import matplotlib.pyplot as plt, numpy as np, pandas as pd
```

```python
x = np.linspace(0, 20, 100)
```

```python
y = x**2
```

```python
figure, axes = plt.subplots(figsize=(8, 8), facecolor='lightskyblue', layout='constrained')
figure.suptitle('Plot of x vs y (y=x^2)')
axes.set(xlabel = 'X')
axes.set_title('Y', loc='left', fontstyle='oblique', fontsize='medium')
axes.set_xlim(0, 20)
axes.set_ylim(0, 400)
axes.plot(x,y, color='orange')
figure.savefig('Task_1.png')
```

```python
import os 
```

```python
def list_files(filepath, filetype):
   paths = []
   for root, dirs, files in os.walk(filepath):
      for file in files:
         if file.lower().endswith(filetype.lower()):
            paths.append(os.path.join(root, file))
   return(paths)
```

```python
my_files_list = list_files('C:\\Users\\lidia\\Documents\\Programming\\DI_course\\DI_bootcamp\\Week5', 'chipo.txt')
```

```python
file = my_files_list[0]
```

```python
data = pd.read_csv(file, sep='\t')
```

```python
data
```

```python
price = data['item_price'].apply(lambda x : float(x[1:]))
```

```python
price
```

```python
quantity = data['quantity']
```

```python
quantity
```

```python
quantity_price, axes = plt.subplots(figsize=(8, 8), facecolor='salmon', layout='constrained')
quantity_price.suptitle('Relationship between Item Price and Quantity Ordered')
axes.set(xlabel = 'Quantity Ordered', ylabel = 'Item Price')
axes.set_xlim(0, 50)
axes.set_ylim(0, 16)
axes.scatter(price, quantity, marker = 'H', color = 'red', edgecolor = 'pink')

quantity_price.savefig('quantity_price.png')
```

```python
items = data.groupby('item_name').count()
```

```python
items_5 = items.sort_values(by='quantity',ascending=False).head(5)
```

```python
hstgr, axes = plt.subplots()
bar_colors = ['yellow', 'orange', 'darkorange', 'red', 'darkred']
axes.set(title = 'Top 5 Items by Quantity Ordered', ylabel='Total Quantities Ordered', xlabel='Item Name')
axes.bar(items_5.index.values, items_5['quantity'], color=bar_colors)
plt.xticks(rotation=45)
hstgr.savefig('Top_5_Items.png')
plt.show()
```
