# Part II

```python
new_properties = data_bh[data_bh['age'] <= 35]
old_properties = data_bh[(data_bh['age'] > 35) & (data_bh['age'] <= 70)]
very_old_properties = data_bh[data_bh['age'] > 70]
```

```python
new_median_price = new_properties['medv'].median()
old_median_price = old_properties['medv'].median()
very_old_median_price = very_old_properties['medv'].median()

plt.figure(figsize=(10, 6))

plt.bar('New', new_median_price, color='blue', label='New Properties')
plt.bar('Old', old_median_price, color='green', label='Old Properties')
plt.bar('Very Old', very_old_median_price, color='red', label='Very Old Properties')

plt.xlabel('Age Group')
plt.ylabel('Median Price')
plt.title('Median Price vs. Property Age Group')
plt.legend()
plt.show()
```

# Part III

```python
bins = [0, 1, 5, 10, 20, max(data_bh['crim'])]
labels = ['0-1', '1-5', '5-10', '10-20', '20+']
data_bh['crim_bin'] = pd.cut(data_bh['crim'], bins=bins, labels=labels)
```

```python
avg_MEDV_per_CRIM_BIN = data_bh.groupby('crim_bin')['medv'].mean()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(avg_MEDV_per_CRIM_BIN, marker='o')
plt.xlabel('Binned Crime Rate')
plt.ylabel('Average Property Price (MEDV)')
plt.title('Neighborhood Crime Rate Impact on Property Prices')
plt.xticks(range(len(labels)), labels)  # Set x-axis ticks to match the crime rate bins
plt.grid(True)
plt.show()
```
