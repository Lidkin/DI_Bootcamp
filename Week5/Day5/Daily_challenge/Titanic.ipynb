```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

```python
sex_counts = data['Sex'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', colors=['blue', 'pink'])
plt.title('Ratio of Male to Female Passengers')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data['Age'], bins=20, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.title('Distribution of Passenger Ages')
plt.show()
```

```python
survival_counts = data['Survived'].value_counts()
plt.figure(figsize=(6, 6))
plt.bar(survival_counts.index, survival_counts.values, color=['red', 'green'])
plt.xticks(survival_counts.index, ['Did not survive', 'Survived'])
plt.xlabel('Survival')
plt.ylabel('Number of Passengers')
plt.title('Passenger Survival Count')
plt.show()
```

```python
survival_by_gender = data.groupby(['Sex', 'Survived']).size().unstack()
survival_by_gender.plot(kind='bar', stacked=True, color=['red', 'green'], figsize=(8, 6))
plt.xticks([0, 1], ['Female', 'Male'], rotation=0)
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.title('Passenger Survival by Gender')
plt.legend(['Did not survive', 'Survived'])
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['Age'], data['Fare'], alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Correlation between Fare and Age')
plt.show()
```

```python
gender_survival_rate = data.groupby('Sex')['Survived'].mean()
```
